
# Introduksjon til `Complex`-klassen

Denne notatboken viser hvordan du bruker `Complex`-klassen fra **Task 1**.
Vi antar at du har en fil `complex.py` i samme mappe med definisjonen `class Complex`.

> Tips: Kjør gjerne hver celle etter tur. Hvis du endrer på `complex.py`, må du restarte kjernen eller bruke `importlib.reload`.


```

# Importer klassen (forutsetter complex.py i samme mappe)
from complex import Complex

# (Valgfritt) verifiser versjon/metadata hvis du har slikt i filen
try:
    __version__  # noqa: F821
except NameError:
    __version__ = "ukjent"
print("Complex-klassen lastet. Versjon:", __version__)
```


## Grunnleggende: konstruksjon og visning (`__str__` og `__repr__`)


```

z = Complex(1, 2)
y = Complex(3, 4)

print("print(z) ->", z)         # bruker __str__
print("print([z, y]) ->", [z, y])  # bruker __repr__
```


## Felt-tilgang: `re` og `im`


```

print("z.re ->", z.re)
print("z.im ->", z.im)
```


## Standardverdier i konstruktør


```

print("Complex() ->", Complex())
print("Complex(5) ->", Complex(5))
```


## Aritmetikk mellom to komplekse tall


```

print("z + y ->", z + y)
print("z - y ->", z - y)
print("z * y ->", z * y)
```


## Aritmetikk mellom komplekst tall og heltall (bilateralt)


```

print("z + 3 ->", z + 3)
print("3 + z ->", 3 + z)
print("z * 3 ->", z * 3)
print("3 * z ->", 3 * z)
```


## Sammenligning: likhet og ulikhet


```

print("z == y ->", z == y)
print("z != y ->", z != y)
# Et tilfelle med likhet
z2 = Complex(1, 2)
print("z == Complex(1, 2) ->", z == z2)
```


## Flere eksempler


```

# Eksempel fra oppgaveteksten
print("Complex(2, -3) * Complex(2, 3) ->", Complex(2, -3) * Complex(2, 3))
```


## (Valgfritt) Bonus: polarform, divisjon, eksponent, konjugasjon, negasjon
Denne seksjonen forsøker å bruke utvidet funksjonalitet dersom du har implementert **Task 5**.
Hvis det ikke er implementert enda, vil vi fange opp feil og fortsette.


```

import math

def try_run(desc, fn):
    try:
        print(desc, "->", fn())
    except Exception as e:
        print(desc, "-> Ikke implementert enda:", type(e).__name__, e)

try_run("Complex(r=1, phi=math.pi/4)", lambda: Complex(r=1, phi=math.pi/4))
try_run("print av Complex(r=1, phi=π/4)", lambda: str(Complex(r=1, phi=math.pi/4)))
try_run("konjugat (z.conj)", lambda: getattr(Complex(r=1, phi=math.pi/4), "conj"))
try_run("negasjon (-z)", lambda: -Complex(r=1, phi=math.pi/4))
try_run("Complex(1,2) / Complex(3,4)", lambda: Complex(1,2) / Complex(3,4))
try_run("Complex(1,2) / 5", lambda: Complex(1,2) / 5)
try_run("5 / Complex(1,2)", lambda: 5 / Complex(1,2))
try_run("Complex(1,2) ** 5", lambda: Complex(1,2) ** 5)
try_run("5 ** Complex(1,2)", lambda: 5 ** Complex(1,2))
```
